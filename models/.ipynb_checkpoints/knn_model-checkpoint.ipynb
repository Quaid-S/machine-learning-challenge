```python
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
```

```python
df = pd.read_csv("../data/exoplanet_data.csv")

df = df.dropna(axis='columns', how='all')

df = df.dropna()
df.head()
```

```python
X = df.drop(['koi_disposition', 'koi_srad_err1', 'koi_steff', 'koi_srad', 'koi_slogg_err2', 'koi_slogg', 'koi_slogg_err1', 'koi_srad_err2', 'koi_tce_plnt_num'], axis=1)
y = df["koi_disposition"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
X_scaler = StandardScaler().fit(X_train.values.reshape(-1, 1))
```

```python
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
train_scores = []
test_scores = []

for k in range(1, 80, 2):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train_scaled, y_train)
    train_score = knn.score(X_train_scaled, y_train)
    test_score = knn.score(X_test_scaled, y_test)
    train_scores.append(train_score)
    test_scores.append(test_score)
    print(f'k: {k}, Train/Test Score: {train_score:.3f}/{test_score:.3f}')
    
```

```python
plt.plot(range(1, 80, 2), train_scores, marker='o')
plt.plot(range(1, 80, 2), test_scores, marker="x")
plt.xlabel('K Neighbors')
plt.ylabel('Testing Acc')
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=15)
knn.fit(X_train_scaled, y_train)
print('k=15 Test Acc: %.3f' % knn.score(X_test_scaled, y_test))
```
